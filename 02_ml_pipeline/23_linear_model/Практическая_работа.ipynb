**Задача 5. Стратегия голосования**

Реализуйте стратегию голосования для предсказания целевой переменной.
Голосование в задаче классификации — это когда несколько моделей выдают свои предикты, и финальным выбирается тот предикт, который предсказали большинство моделей.

Для реализации этой стратегии проделайте следующее:

- сохраните предсказания каждой из моделей (случайный лес, многослойный персептрон, логистическая регрессия) для тестовой выборки в датафрейм `pred_df`;
- в четвёртую колонку `target` положите тот класс, который предсказало большинство классификаторов; например, если в строке были значения `high, medium, medium`, в `target` нужно положить `medium`;

     если в строке три разных класса (`high, medium, low`), придумайте свою стратегию по выбору значения; самая простая стратегия — выбрать рандомно одно значение из трёх;

- посчитайте точность предсказания с помощью голосования; выведите значения метрик accuracy и confusion_matrix.

Добейтесь значения точности > 78%.

```python
# Ваш код здесь
pred_lgr = lgr.predict(X_test)
pred_mlp = mlp.predict(X_test)
pred_rf = rf.predict(X_test)

pred_df = pd.DataFrame({
    'lgr': pred_lgr,
    'mlp': pred_mlp,
    'rf': pred_rf
})
pred_df.head()
```

```python
def voting_strategy(row):
    votes = [row['lgr'], row['mlp'], row['rf']]
    
    vote_counts = {}
    for vote in votes:
        if vote in vote_counts:
            vote_counts[vote] += 1
        else:
            vote_counts[vote] = 1
    
    max_votes = 0
    for count in vote_counts.values():
        if count > max_votes:
            max_votes = count
    
    if max_votes >= 2:
        for vote, count in vote_counts.items():
            if count == max_votes:
                return vote
    
    return row['mlp']

pred_df['target'] = pred_df.apply(voting_strategy, axis=1)

voting_accuracy = accuracy_score(y_test, pred_df['target'])
print(f'Voting Accuracy: {voting_accuracy}')
print(confusion_matrix(y_test, pred_df['target']))

print(f'LogRegr Accuracy: {accuracy_score(y_test, pred_df["lgr"])}')
print(f'MLP Accuracy:     {accuracy_score(y_test, pred_df["mlp"])}')
print(f'RF Accuracy:      {accuracy_score(y_test, pred_df["rf"])}')
print(f'Voting Accuracy:  {voting_accuracy:.4f}')

print('\nПерепробовал несколько стратегий:')
print('- рандомный выбор')
print('- выбор конкретного класса из трех') 
print('- выбор взвешенный на основе точности моделей')
print('- выбор конкретной модели')
print('Выше 77.89% результат получить не удалось.')
```
